```python
import random
from IPython.display import clear_output
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}
chips = 1000

print("------------------WELCOME TO THE GAME OF BLACKJACK!------------------")

def shuffling(deck):
    random.shuffle(deck)

def bet(chip):
    while True:
        try:
            amt = int(input("Enter your bet : "))
        except:
            print("Not a valid input. Please try again.")
        else:
            if(amt <= chip):            
                return amt
                break
            else:
                print("You don't have that many chips.")
            
while True:
    deck = []
    flag = 1
    for suit in suits:
        for rank in ranks:
            deck.append((suit, rank))
    player_val = 0
    dealer_val = 0
    shuffling(deck)
    player_bet = bet(chips)    
    player_hand = []
    dealer_hand = []
    for i in range(2):
        player_hand.append(deck.pop())
        dealer_hand.append(deck.pop())
    print("player has:")
    for suit, rank in player_hand:
        print(f"{rank} of {suit}")
        player_val += values[rank]
    a = 0
    print(f"Player's hand value : {player_val}")
    for suit, rank in dealer_hand:
        if a == 0:
            print(f"One of the dealer's cards is : {rank} of {suit}.")
            a += 1
        dealer_val += values[rank]
    input("Press any button to continue.....")
    clear_output()
    while True:
        print(f"Current player's hand value : {player_val}")
        while True:
            choice = str(input("Press 'h' to hit and 's' to stay : "))
            if(choice in 'hs'):            
                break
            else:
                print("You have entered an invalid choice. Please try again.")
        if(choice == 's'):
            break
        else:
            hand = deck.pop()
            suit, rank = hand
            player_hand.append(hand)
            print(f"Card picked up by player : {rank} of {suit}")            
            player_val += values[player_hand[-1][1]]
            print(f"Current player's hand value : {player_val}")
            if(player_val > 21):
                print("Player loses the round.")
                chips -= player_bet
                flag = 0
                break
            elif(player_val == 21):
                chips += player_bet * 1.5
                print("Player wins the round!")
                flag = 0
                break
            input("Press any button to continue.....")
            clear_output()
    if(flag == 1):
        suit, rank = dealer_hand[1]
        print(f"Dealer's other card is : {rank} of {suit}")
        while(dealer_val <= 17):
            hand = deck.pop()            
            suit, rank = hand
            dealer_hand.append(hand)
            print(f"Card picked up by the dealer : {rank} of {suit}")
            dealer_val += values[dealer_hand[-1][1]]
        print(f"Current dealer value : {dealer_val}")
        if(dealer_val > 21):
            chips += player_bet
            print("Player wins the round!")
        elif(dealer_val < player_val):
            chips += player_bet
            print("Player wins the round!")
        elif(dealer_val > player_val):
            print("Player loses the round.")
            chips -= player_bet
        else:
            print("It is a draw.")
            print(f"Chips left : {chips}")
    print(f"Chips left = {chips}")
    if(str(input("Enter 'n' to quit and any other alphabet to continue : ")) == 'n'):
        break
    clear_output()
    
print("--------------------COME BACK AGAIN!--------------------")
```
